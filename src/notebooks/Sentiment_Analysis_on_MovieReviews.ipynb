## 1. What is the nature of the leading opinions in service review on computer games?
#### Objective:
Determine the general sentiment of the reviews and identify the main topics or themes that emerge from the opinions.

```python
import nltk
from nltk.corpus import stopwords
import string

# Download stopwords if not already downloaded
nltk.download('stopwords')

# Function to clean review text
def clean_review(text):
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text
df['cleaned_review'] = df['review'].apply(clean_review)
```

```python
df[['review', 'cleaned_review']].head()
```

```python
# Convert timestamps to datetime
df['timestamp_created'] = pd.to_datetime(df['timestamp_created'], unit='s')
df['timestamp_updated'] = pd.to_datetime(df['timestamp_updated'], unit='s')

df[['timestamp_created', 'timestamp_updated']].head()
```

## Overview of VADER

### VADER Sentiment Analyzer:

- **Tool:** VADER is a lexicon and rule-based sentiment analysis tool specifically designed to analyze sentiments expressed in text, particularly suited for social media and other informal text.
- **Components:** It uses a pre-built lexicon of sentiment-related words and rules to compute sentiment scores. It categorizes text into positive, negative, or neutral based on these scores.
- **Output:** VADER provides a compound score that summarizes the overall sentiment of the text, along with separate scores for positive, negative, and neutral sentiments.

### How It Works:

#### Sentiment Scores:

- **Positive Score:** Measures the proportion of positive sentiment in the text.
- **Negative Score:** Measures the proportion of negative sentiment in the text.
- **Neutral Score:** Measures the proportion of neutral sentiment in the text.
- **Compound Score:** A normalized score that combines the positive, negative, and neutral scores into a single value ranging from -1 (most negative) to +1 (most positive). This score is used to classify sentiment as positive, neutral, or negative.


```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()


# Function to classify sentiment based on VADER scores
def classify_sentiment(text):
    score = sia.polarity_scores(text)
    compound_score = score['compound']
    if compound_score >= 0.05:
        return 'positive'
    elif compound_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment'] = df['cleaned_review'].apply(classify_sentiment)

df[['cleaned_review', 'sentiment']].head()
```

```python
## Plot
import matplotlib.pyplot as plt

# Calculate sentiment distribution
sentiment_counts = df['sentiment'].value_counts()
sentiment_labels = sentiment_counts.index
sentiment_sizes = sentiment_counts.values

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sentiment_sizes, labels=sentiment_labels, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'], startangle=140)
plt.title('Sentiment Distribution of Game Reviews')
plt.show()
```

#### Interpretation
1. **Neutral Dominance:** A large majority of the reviews are neutral (71.1%). This might indicate that many users didn't have strong opinions or that the reviews were mostly factual rather than emotional.
2. **Positive Sentiments:** Positive reviews constitute a significant portion (25.2%). This suggests that a notable number of users had a favorable view of the games.
3. **Negative Sentiments:** Negative reviews are quite low (3.7%). This could indicate that negative experiences are less frequently shared or that the overall user experience is generally positive.

## 2. Which elements of games are the most frequently recalled in the reviews?
#### Objective:
Identify and analyze the main themes or topics that are frequently mentioned in the game reviews. This can reveal what aspects of the games (such as graphics, gameplay, story) are most often highlighted by reviewers.

### Steps for Topic Modeling
#### Preprocess the Text Data:

1. Tokenize the cleaned reviews.
Remove stopwords and non-alphanumeric tokens.
2. Vectorize the Text:
Convert the text data into numerical form using techniques like TF-IDF (Term Frequency-Inverse Document Frequency).
3. Apply Topic Modeling:
Use a technique like LDA (Latent Dirichlet Allocation) to extract topics from the text data.
4. Interpret the Topics:
Examine the top words associated with each topic to understand what themes they represent.


```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

# Download required NLTK resources
nltk.download('punkt')

# Function to tokenize and clean text
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word.isalnum()]
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.lower() not in stop_words]
    return ' '.join(tokens)

df['processed_review'] = df['cleaned_review'].apply(preprocess_text)

df[['processed_review']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000) 
X = tfidf_vectorizer.fit_transform(df['processed_review'])
print(X.shape)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# Initialize LDA model
num_topics = 5  # Number of topics to extract
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(X)

# Display the top words for each topic
feature_names = tfidf_vectorizer.get_feature_names_out()

def print_top_words(model, feature_names, n_words=10):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic #{topic_idx + 1}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-n_words - 1:-1]]))

print_top_words(lda, feature_names)
```

## Interpretation:
